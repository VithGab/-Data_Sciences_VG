Desafio 1.3: Tendência de Acessos ao Site

Cenário: A equipe de produto quer monitorar a evolução diária do número de acessos ao site nos últimos 30 dias para identificar padrões ou quedas.

```python
# Dados para o Desafio 1.3
datas = pd.date_range(start='2024-03-01', periods=30)
np.random.seed(42)
acessos = np.random.randint(1000, 5000, 30)
# Simular uma queda nos últimos dias
acessos[-5:] = np.random.randint(500, 1500, 5)
df_acessos = pd.DataFrame({'Data': datas, 'Acessos': acessos})
print("\nDataFrame para Desafio 1.3 (primeiras 5 linhas):")
print(df_acessos.head())
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: [SUA RESPOSTA]
# 2. Justificativa (Cleveland-McGill, Gestalt): [Grafico em linha ajuda na maior percpção, já que o mesmo diminui o esforço cognitivo(Cleveland)
# Além disso há três principios, continuidade, conectando pontos e linhas à referência, proximidade, o agrupamento das data no eixo x ajuda o cerebro a
# entender os intervalos de tempo, Destino comum, o cerebro identifica padrões com a forma tomada pela linha.]

# Seu código Python para o gráfico:
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

plt.plot(df_acessos['Data'], df_acessos['Acessos'],
         marker='o', linestyle='-', color='purple', linewidth=2, markersize=5)

plt.ylim(0, df_acessos['Acessos'].max() * 1.1)
plt.title('Evolução Diária de Acessos - Março/2024', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Total de Acessos', fontsize=12)
plt.xticks(rotation=45)

plt.axvspan(df_acessos['Data'].iloc[-5], df_acessos['Data'].iloc[-1],
            color='red', alpha=0.1, label='Período de Queda')

plt.legend()
plt.tight_layout()
plt.show()
```

Desafio 1.4: Relação entre Investimento em Marketing e Vendas

Cenário: A equipe de finanças quer entender se há uma correlação entre o investimento mensal em marketing e o volume de vendas gerado no mesmo mês.

```python
# Dados para o Desafio 1.4
np.random.seed(42)
investimento_marketing = np.random.randint(10000, 50000, 12)
vendas_mensais = investimento_marketing * 5 + np.random.randint(-5000, 5000, 12)
df_marketing_vendas = pd.DataFrame({
    'Investimento_Marketing': investimento_marketing,
    'Vendas_Mensais': vendas_mensais
})
print("\nDataFrame para Desafio 1.4 (primeiras 5 linhas):")
print(df_marketing_vendas.head())
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: [Scatter plot]
# 2. Justificativa (Cleveland-McGill, Gestalt): [O eixo y ajuda a identificar os outliers e as metricas comparativas de proporcionalidade, no caso proporção direta com poucos outliers
# gestalt uso das cores gera uma Semelhança, ao usar cores diferentes que ajudam a acomparar os dados e suas proporções, permitindo isolar grupos e analisar as metricas.]

# Seu código Python para o gráfico:
cores_roxas = ['#4B0082', '#6A5ACD', '#8A2BE2', '#9370DB', '#B19CD9']
lista_cores = [cores_roxas[i % len(cores_roxas)] for i in range(len(df_marketing_vendas))]

df_ordenado = df_marketing_vendas.sort_values(by='Investimento_Marketing')
plt.figure(figsize=(10, 6))
plt.scatter(df_marketing_vendas['Investimento_Marketing'], df_marketing_vendas['Vendas_Mensais'],
            color=lista_cores, label='Dados Mensais', zorder=5)

plt.plot(df_ordenado['Investimento_Marketing'], df_ordenado['Vendas_Mensais'],
         color='black', linestyle='-', alpha=0.6, label='Conexão')

plt.title('Investimento em Marketing vs Vendas Mensais')
plt.xlabel('Investimento (R$)')
plt.ylabel('Vendas (R$)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

Desafio 1.5: Composição de Mercado por Categoria de Produto

Cenário: Um analista de mercado quer visualizar a proporção de cada categoria de produto no total de vendas da empresa para entender a dominância de certas categorias.

```python
# Dados para o Desafio 1.5
dados_categorias = {
    'Categoria': ['Eletrônicos', 'Vestuário', 'Alimentos', 'Livros', 'Serviços'],
    'Vendas_Milhoes': [35.0, 20.0, 15.0, 10.0, 5.0]
}
df_categorias = pd.DataFrame(dados_categorias)
print("\nDataFrame para Desafio 1.5:")
print(df_categorias)
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: [Donut chart]
# 2. Justificativa (Cleveland-McGill, Gestalt): [Gestalt, O circulo é o fechamento, Já que
# nele temos a similariedade e proximidade, essas três medidas de parametros ajudam a
#reconhecer a categorização e a divisão dos produtos]

# Seu código Python para o gráfico:
cores_roxas = ['#4B0082', '#6A5ACD', '#8A2BE2', '#9370DB', '#B19CD9']

plt.figure(figsize=(8, 6))
plt.pie(df_categorias['Vendas_Milhoes'], labels=df_categorias['Categoria'],
        autopct='%1.1f%%', startangle=140, colors=cores_roxas, pctdistance=0.85)

centro_circulo = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centro_circulo)

plt.title('Vendas por Categoria (Milhões)', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Desafio 2.1: Vendas Trimestrais com Eixo Y Enganoso

Cenário: Um gerente de vendas recebeu este gráfico e está preocupado com uma suposta queda drástica nas vendas do Q4.

```python
# Dados para o Desafio 2.1
dados_trimestrais = {
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [1000, 1050, 1020, 980]
}
df_trimestral = pd.DataFrame(dados_trimestrais)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='lightcoral')
ax.set_ylim(bottom=950) # Eixo Y começando alto
ax.set_title('Vendas Trimestrais - Último Ano', fontsize=14, fontweight='bold')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# 1. Problemas identificados:
#    a) Eixo Y Enganoso, eixo Y começando em 950 distorce visualmente as diferenças nas vendas,
#    b) Falta dos valores correspondentes exatos aos valores atribuidos a cada trimestre.
#       A adição dos rótulos facilitam a visualização dos reais valores, acarretando nas tomadas de decisões.
#    c) Cores e Grid Excessivo, a cor 'lightcoral' e linhas de grade fortes podem ser consideradas chartjunk.

# 2. Seu código Python refatorado:
plt.figure(figsize=(7, 5))

ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='purple')

ax.set_ylim(0, df_trimestral['Vendas'].max() * 1.15)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=11, fontweight='bold')

ax.set_title('Vendas Trimestrais - Último Ano (Escala Corrigida)', fontsize=14, fontweight='bold')
ax.set_xlabel('Trimestre', fontsize=12)
ax.set_ylabel('Vendas (Unidades)', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

Desafio 2.2: Comparação de Produtos com Gráfico de Pizza Lotado

Cenário: A equipe de produto quer ver a proporção de vendas de 8 produtos diferentes, mas o gráfico atual está confuso.

```python
# Dados para o Desafio 2.2
dados_produtos = {
    'Produto': [f'Produto {i}' for i in range(1, 9)],
    'Vendas': [150, 120, 80, 70, 60, 50, 40, 30]
}
df_produtos = pd.DataFrame(dados_produtos)

plt.figure(figsize=(8, 8))
plt.pie(df_produtos['Vendas'], labels=df_produtos['Produto'], autopct='%1.1f%%', startangle=90)
plt.title('Proporção de Vendas por Produto', fontsize=14, fontweight='bold')
plt.show()
```

```python
# 1. Problemas identificados:
#    a) [Poluição visual]
#    b) [Exesso de partições]
#    c) [Baixo contraste]

# 2. Seu código Python refatorado:
cores_roxas = ['#4B0082', '#6A5ACD', '#8A2BE2', '#9370DB', '#B19CD9']

plt.figure(figsize=(10, 6))
bars = plt.barh(df_produtos['Produto'], df_produtos['Vendas'], color=cores_roxas)
plt.xlabel('Volume de Vendas', fontsize=12)
plt.title('Ranking de Vendas por Produto', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
sns.despine()
for bar in bars:
    plt.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

Desafio 2.4: Gráfico de Linhas com Chartjunk Excessivo

Cenário: Um analista de dados criou este gráfico para mostrar a evolução de duas métricas ao longo do tempo, mas ele está muito poluído.

```python
# Dados para o Desafio 2.4
datas_mes = pd.date_range(start='2023-01-01', periods=12, freq='M')
np.random.seed(42)
metrica_a = np.random.randint(50, 150, 12).cumsum()
metrica_b = np.random.randint(40, 130, 12).cumsum()

df_metricas = pd.DataFrame({
    'Data': datas_mes,
    'Metrica_A': metrica_a,
    'Metrica_B': metrica_b
})

plt.figure(figsize=(10, 6))
plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A', marker='o', linestyle='-', linewidth=2, color='blue')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B', marker='o', linestyle='-', linewidth=2, color='black')
#

plt.grid(True)
plt.title('Evolução das Métricas A e B', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Problemas identificados:
#    a) [As cores não ajudavam na visualização do gráfico]
#    b) [As linhas não seguiam o mesmo padrão de estilo]
#    c) [as datas estavam confusas]

# 2. Seu código Python refatorado:
# O código foi alterado no própio enúnciado
```

Desafio 2.5: Comparação de Desempenho com Gráfico de Bolhas Confuso

Cenário: Um gerente de produto quer comparar o desempenho de diferentes funcionalidades (Funcionalidade 1 a 5) com base em duas métricas: 'Uso' e 'Satisfação'. O tamanho da bolha representa o 'Impacto'.

```python
# Dados para o Desafio 2.5
dados_funcionalidades = {
    'Funcionalidade': [f'Func {i}' for i in range(1, 6)],
    'Uso': [0.8, 0.6, 0.9, 0.4, 0.7],
    'Satisfacao': [0.9, 0.7, 0.8, 0.5, 0.6],
    'Impacto': [100, 50, 120, 30, 80]
}
df_funcionalidades = pd.DataFrame(dados_funcionalidades)

plt.figure(figsize=(8, 8))
plt.scatter(x=df_funcionalidades['Uso'], y=df_funcionalidades['Satisfacao'],
            s=df_funcionalidades['Impacto']*5, # Multiplicar para bolhas maiores
            alpha=0.7,
            c=np.random.rand(len(df_funcionalidades), 3)) # Cores aleatórias

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'], row['Funcionalidade'], fontsize=9, ha='center', va='center')

plt.title('Desempenho das Funcionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Uso (0-1)', fontsize=12)
plt.ylabel('Satisfação (0-1)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

```python
# 1. Problemas identificados:
#    a) [Legibilidade dos rótulos das bolhas]
#    b) [Paleta de cores, muitas cores diferentes]
#    c) [Tamanhos das bolhas, a melhor prática é usar o tamnho
#        das bolhas para indicar a dimenção do impacto]

# 2. Seu código Python refatorado:
cores_roxas = ['#B19CD9', '#9370DB', '#8A2BE2', '#6A5ACD', '#4B0082']

plt.figure(figsize=(9, 7))

scatter = plt.scatter(x=df_funcionalidades['Uso'], y=df_funcionalidades['Satisfacao'],
                      s=df_funcionalidades['Impacto']*15, alpha=0.7, c=cores_roxas,
                      edgecolors='white', linewidth=1.5)

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'] + 0.03, row['Funcionalidade'],
             fontsize=10, ha='center', fontweight='bold')

plt.title('Desempenho das Funcionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Uso (0-1)', fontsize=12)
plt.ylabel('Satisfação (0-1)', fontsize=12)
plt.xlim(0.3, 1.0)
plt.ylim(0.4, 1.0)
plt.grid(True, linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()
```

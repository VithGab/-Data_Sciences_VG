```python
import ucimlrepo
import pandas as pd
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
adult = fetch_ucirepo(id=2)

# data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# metadata
print(adult.metadata)

# variable information
print(adult.variables)
```

```python
full_df = pd.concat( [X, y], axis=1)
```

```python
full_df.head()
```

```python
full_df.columns
```

```python
full_df = full_df.rename(columns = {'hours-per-week':'hours_per_week',
                                    'capital - gain':'capital_gain',
                                    'capital - loss':'capital_loss',
                                    'native - country':'native_country'
                                    })
```

```python
full_df["education"].value_counts()
```

```python
full_df["education"] = full_df["education"].replace({"Preschool": "preschool",
                                                     'HS-grad': 'hs-grad',
                                                     'Some-college': 'some-college',
                                                     'Bachelors': 'bachelors',
                                                     'Masters': 'masters',
                                                     'Prof-school': 'prof-school',
                                                     'Assoc-voc': 'assoc-voc',
                                                     'Doctorate': 'doctorate',
                                                     'Assoc-acdm':'assoc-acdm'})
```

```python
full_df["education"] = full_df["education"].fillna("unknown")
```

**Simulação de Monte Carlo**

Como funciona? (O passo a passo)
1. Defina as variáveis: Escolha o que você quer medir (ex: lucro de um projeto).

2. Identifique as incertezas: Em vez de usar um número fixo (ex: "o custo será R$ 100"), você define um intervalo (ex: "o custo varia entre R$ 80 e R$ 120").

3. Rode a roleta: O computador escolhe números aleatórios dentro desses intervalos e calcula o resultado.

4. Repita (muito!): Fazemos isso 10.000 vezes ou mais.

5. Analise o resultado: O resultado final não é um número único, mas uma distribuição de probabilidades.

**Simulando seu Lucro**
Imagine que você está lançando um produto:

Vendas esperadas: Média de 1.000 unidades (mas pode variar para mais ou menos).

Preço de venda: Fixo em R$ 50.

Custo de produção: Varia entre R$ 20 e R$ 35 por unidade.

Vou rodar essa simulação 10.000 vezes para ver qual é a chance de termos um lucro específico.

```python
import numpy as np
import matplotlib.pyplot as plt

N_SIMULACOES = 10000

# --- ESTRATÉGIA A ---
vendas_a = np.random.normal(500, 50, N_SIMULACOES)
custo_u_a = np.random.uniform(600, 700, N_SIMULACOES)
lucro_a = (vendas_a * 1200) - (vendas_a * custo_u_a) - 50000

# --- ESTRATÉGIA B ---
vendas_b = np.random.normal(800, 250, N_SIMULACOES)
custo_u_b = np.random.uniform(400, 800, N_SIMULACOES)
lucro_b = (vendas_b * 1000) - (vendas_b * custo_u_b) - 150000

# --- CÁLCULOS ---
def analise(lucro, nome):
    media = np.mean(lucro)
    prob_prejuizo = (np.sum(lucro < 0) / N_SIMULACOES) * 100
    print(f"Estratégia {nome}: Média R$ {media:.2f} | Prob. Prejuízo: {prob_prejuizo:.2f}%")

analise(lucro_a, "A")
analise(lucro_b, "B")

# --- VISUALIZAÇÃO ---
plt.figure(figsize=(10, 6))
plt.hist(lucro_a, bins=50, alpha=0.5, label='Estratégia A', color='blue')
plt.hist(lucro_b, bins=50, alpha=0.5, label='Estratégia B', color='purple')
plt.axvline(0, color='red', linestyle='--', label='Ponto de Equilíbrio (R$ 0)')

plt.title('Comparação de Lucro: Estratégia A vs B')
plt.xlabel('Lucro (R$)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```
